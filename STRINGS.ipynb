## ✦ O que é uma String em Python?

✧ No mundo da programação, uma string é uma sequência de caracteres.
Isso significa que palavras, frases, números, emojis e até símbolos especiais são tratados como strings.

Em Python, podemos criar strings usando aspas simples `' '` ou aspas duplas `" "`.

**Exemplos:**

```python
"Olá, mundo!" # Frase
'STEAM' # Palavra
"2025" # Número como texto
'!@#$%&*?' # Caracteres especiais
```

✧ **LEMBRE-SE:** Mesmo que uma string tenha só números, ela não é um número, ela é um texto.
```python
idade = 16 # Isso é um número
idade = "16" # Isso é uma string
```





## ✦ Para que serve uma String?

✧ Strings são usadas para armazenar, exibir e processar textos em programas.

**Principais usos:**


*   **Armazenar informações:** Salvar nomes, endereços, textos e qualquer dado de texto.

```python
nome = "Humberto"
mensagem = "Bem-vindo!"
```

*   **Exibir informações:** Tudo que o Python imprime no terminal para o usuário, geralmente, é uma string.

```python
print("Olá, mundo!")  # Exibe texto na tela.
```

*   **Receber dados do usuário:** Quando alguém digita algo em um programa, o valor que chega é sempre string!

```python
nome = input("Qual seu nome? ")
```
*   **Manipular textos:** É possível modificar partes de uma string, contar caracteres, buscar palavras, cortar pedaços, corrigir erros de digitação e formatar para impressão.

```python
texto = "   Olá, mundo!   "
limpo = texto.strip()  # Remove os espaços
print(limpo)           # Olá, mundo!
```

*   **Processar dados:** Muitos dados em sistemas são recebidos como texto e precisam ser tratados antes de usar, como datas, nomes, senhas e respostas de formulários.

```python
data = "05/05/2025"
partes = data.split("/")
print(partes)  # ['05', '05', '2025']
```

*   **Comunicação com o usuário:** Strings são essenciais para gerar mensagens, relatórios, erros, notificações e alertas.

```python
print("Erro: senha inválida.")
```










## ✦ Relação entre Strings e Listas

✧ Em Python, strings são muito parecidas com listas! Cada caractere ocupa
uma posição específica, que chamamos de índice. Essa posição começa do número 0, e assim como nas listas, podemos acessar elementos diretamente pelo número do índice.


**Exemplo:** palavra = "Python"

Python att.png


✧ Porém, diferente das listas, as strings são **IMUTÁVEIS**: isso significa que não podemos
alterar um caractere diretamente pelo índice. Sempre que precisamos modificar uma string,
criamos uma nova string usando operações ou métodos.


```python
palavra = "Python"
print(palavra[0])  # P
print(palavra[1])  # y
```

## ✦ Operadores com Strings

✧ Em Python, podemos usar operadores para trabalhar com strings, facilitando a construção e manipulação de textos.

---


### **Operador de Concatenação (+)**

O operador `+` é usado para juntar (concatenar) duas ou mais strings.  Quando usamos `+`, as strings são coladas uma na outra. **Exemplo:**

```python
nome = "Ana"
sobrenome = "Silva"
print(nome + " " + sobrenome)  # Resultado: Ana Silva
```

---

### **Operador de Repetição (*)**

O operador `*` permite repetir uma string várias vezes.
A string será repetida o número de vezes indicado pelo número inteiro. **Exemplo:**

```python
risada = "ha"
print(risada * 3)  # Resultado: hahaha
```


## ✦ Fatiamento de Strings (Slicing)

✧ Com fatiamento, conseguimos pegar pedaços da string.

---
**Sintaxe:**

```
string[início:fim:passo]
```

*   início → índice de onde começa
*   fim → até onde vai (não inclui o índice final!)
*   passo → de quantos em quantos pula
---
**Exemplo:**

SLICING.png

````python
texto = "Programação"
print(texto[0:4]) # Prog
print (texto[:5]) # Progr
print (texto[3:]) # gramação
print (texto[::-1]) # oãçamargorP
````

## ✦ Análise de Strings

✧ São funções que te ajudam a examinar ou buscar informações dentro da string. Os mais comuns para analisar strings são:

| Método      | O que faz |
| :------:    | :-----:   |
| **len():**  | Conta o número total de caracteres na string    |
| **count():**| Conta quantas vezes um caractere ou trecho aparece  |
| **find():** | Retorna a posição da primeira ocorrência de um trecho  |
| **in:**     |Verifica se um trecho existe dentro da string (retorna True/False)  |



````python
texto = "banana"
print(len(texto))           # Tamanho: 6
print(texto.count("a"))     # Quantas vezes 'a' aparece: 3
print(texto.find("na"))     # Posição da primeira ocorrência de 'na': 2
print("ban" in texto)       # Verifica se existe 'ban': True
````

## ✦ Transformação de Strings

✧ Em Python, strings podem ser modificadas através de métodos. Esses métodos permitem alterar o conteúdo ou a forma da string sem mudar o objeto original (strings são imutáveis, então esses métodos sempre retornam uma nova string!).

Abaixo estão os principais métodos de transformação:


| Método              | O que faz |
| :------:            | :-----:   |
| **strip()**         | Remove espaços em branco do início e do fim da string    |
| **upper():**        | Converte todos os caracteres para maiúsculas  |
| **lower():**        | Converte todos os caracteres para minúsculas  |
| **capitalize():**   | Coloca a primeira letra em maiúscula e o restante em minúsculo  |
| **title():**        | Coloca a primeira letra de cada palavra em maiúscula |
| **replace():**      | Substitui uma parte da string por outra |
| **rstrip():**        | 	Remove espaços em branco à direita (final da string)
| **lstrip():**        | 	Remove espaços em branco à esquerda (início da string)







````python
texto = "  Olá, mundo!  "
print(texto.strip())         # Remove espaços laterais
print(texto.upper())         # Tudo maiúsculo
print(texto.lower())         # Tudo minúsculo
print(texto.strip().capitalize())    # Primeira letra maiúscula e sem espaços laterais
print(texto.title())         # Cada palavra com inicial maiúscula
print(texto.replace("mundo", "Python"))  # Substitui palavra
print(texto.rstrip())        # Remove espaços da direita
print(texto.lstrip())        # Remove espaços da esquerda
````

## ✦ Divisão de Strings `split()`

✧ Com `split()`, podemos dividir uma string em partes, gerando uma lista. **Exemplo:**

```python
frase = "Python é divertido"
print(frase.split()) # ['Python', 'é', 'divertido']
```

Também podemos passar um separador:



```python
data = "05/05/2025"
print(data.split("/")) # ['05', '05', '2025']
```

## ✦ Junção de Strings `join()`

✧ O método `join()` junta uma lista de strings em uma única string.
**Exemplo:**

```python
lista = ['Python', 'é', 'legal']
frase = " ".join(lista)
print(frase)  # Python é legal
```


Também é muito usado para formatar datas ou separar valores com vírgulas.



```python
numeros = ['05', '05', '2025']
data_formatada = "/".join(numeros)
print(data_formatada)  # 05/05/2025

```
